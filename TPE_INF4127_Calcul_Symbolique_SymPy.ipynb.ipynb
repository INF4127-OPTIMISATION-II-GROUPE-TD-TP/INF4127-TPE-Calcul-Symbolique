## BLOC 2: Cr√©ation des Symboles Math√©matiques

```python
# BLOC 2: Cr√©ation des symboles
print(" CR√âATION DES SYMBOLES MATH√âMATIQUES")

# Cr√©ation de symboles simples
x, y, z = symbols('x y z')
print(f" Symboles cr√©√©s: x = {x}, y = {y}, z = {z}")
print(f" Type de x: {type(x)}")

# Cr√©ation de symboles avec propri√©t√©s sp√©cifiques
a, b, c = symbols('a b c', real=True)  # Nombres r√©els
n, m = symbols('n m', integer=True)    # Nombres entiers
print(f" Symboles r√©els: a = {a}, b = {b}, c = {c}")
print(f" Symboles entiers: n = {n}, m = {m}")

# Test d'expression simple
expr_test = x**2 + 2*x + 1
print(f" Expression test: {expr_test}")

# V√©rification que les symboles fonctionnent
print(f" Substitution x=3: {expr_test.subs(x, 3)}")
```

## BLOC 3: Calcul Exact avec Fractions et Constantes

```python
# BLOC 3: Calcul exact vs approximation num√©rique
print(" CALCUL EXACT VS APPROXIMATION NUM√âRIQUE")

# Expression avec fractions exactes
expr_exact = (x + 1)**2 / Rational(2, 3) + sqrt(2)
print(f" Expression exacte: {expr_exact}")

# √âvaluation num√©rique avec haute pr√©cision
exact_value = expr_exact.subs(x, 1).evalf(15)
print(f" Valeur exacte en x=1: {exact_value}")

# Comparaison avec l'approximation flottante
approx_float = (1 + 1)**2 / (2/3) + 1.4142135623730951
print(f" Approximation flottante: {approx_float}")

# Calcul de l'erreur
error = abs(exact_value - approx_float)
print(f" Erreur d'approximation: {error}")

# Test avec œÄ
print(f"\n Test avec œÄ:")
print(f"œÄ exact (SymPy): {pi}")
print(f"œÄ approxim√© (math): {np.pi}")
print(f"Diff√©rence: {abs(pi.evalf(15) - np.pi)}")

# Test avec une autre expression
expr_complexe = (sqrt(2) + sqrt(3)) / Rational(1, 2)
print(f"\n Expression complexe: {expr_complexe}")
print(f"Valeur num√©rique: {expr_complexe.evalf(10)}")
```

## BLOC 4: Manipulation Alg√©brique des Expressions

```python
# BLOC 4: Manipulation alg√©brique
print(" MANIPULATION ALG√âBRIQUE")

# Exemple 1: D√©veloppement
expr_dev = (x + 1)**3
developpe = expand(expr_dev)
print(f" D√©veloppement de (x+1)¬≥: {developpe}")

# Exemple 2: Factorisation
expr_fac = x**3 + 3*x**2 + 3*x + 1
factorise = factor(expr_fac)
print(f" Factorisation de x¬≥+3x¬≤+3x+1: {factorise}")

# Exemple 3: Simplification
expr_simp = (x**2 - 1)/(x - 1)
simplifie = simplify(expr_simp)
print(f" Simplification de (x¬≤-1)/(x-1): {simplifie}")

# Exemple 4: Collecte de termes
expr_collect = x*y + x - 3 + 2*x**2 - x**2*y
collecte_x = sp.collect(expr_collect, x)
collecte_y = sp.collect(expr_collect, y)
print(f" Collecte par rapport √† x: {collecte_x}")
print(f" Collecte par rapport √† y: {collecte_y}")

# Exemple 5: Test avec une expression plus complexe
expr_complexe = (x**2 - y**2) / (x - y) + (x**3 - 1) / (x - 1)
simplifie_complexe = simplify(expr_complexe)
print(f"\n Expression complexe: {expr_complexe}")
print(f" Simplifi√©e: {simplifie_complexe}")
```

## BLOC 5: Calcul Diff√©rentiel (D√©riv√©es)

```python
# BLOC 5: Calcul diff√©rentiel
print(" CALCUL DIFF√âRENTIEL - D√âRIV√âES")

# D√©riv√©e simple
f = sin(x)*x
f_prime = diff(f, x)
print(f" Fonction: f(x) = {f}")
print(f" D√©riv√©e: f'(x) = {f_prime}")

# √âvaluation en un point
x_point = 1.0
valeur_exacte = f_prime.subs(x, x_point).evalf()
print(f" f'({x_point}) exact = {valeur_exacte}")

# D√©riv√©es d'ordre sup√©rieur
f_seconde = diff(f, x, 2)
f_troisieme = diff(f, x, 3)
print(f" D√©riv√©e seconde: f''(x) = {f_seconde}")
print(f" D√©riv√©e troisi√®me: f'''(x) = {f_troisieme}")

# Test avec une fonction plus complexe
f_complexe = exp(x) * cos(x)
f_complexe_prime = diff(f_complexe, x)
print(f"\n Fonction complexe: {f_complexe}")
print(f" D√©riv√©e: {f_complexe_prime}")

# Test avec r√®gle de cha√Æne
f_chain = sin(x**2 + 1)
f_chain_prime = diff(f_chain, x)
print(f"\n  Fonction avec composition: {f_chain}")
print(f" D√©riv√©e (r√®gle de cha√Æne): {f_chain_prime}")

# V√©rification de la r√®gle de produit
print(f"\n V√©rification r√®gle de produit:")
print(f"d/dx [sin(x)*x] = cos(x)*x + sin(x)*1")
print(f"R√©sultat SymPy: {f_prime}")
```

## BLOC 6: Gradient et D√©riv√©es Partielles (Multivari√©)

```python
# BLOC 6: Gradient et d√©riv√©es partielles
print(" GRADIENT ET D√âRIV√âES PARTIELLES")

# Fonction de deux variables
f_multi = x**2 * y**3 + cos(x)
print(f" Fonction multivari√©e: f(x,y) = {f_multi}")

# D√©riv√©es partielles
df_dx = diff(f_multi, x)
df_dy = diff(f_multi, y)
print(f" D√©riv√©e partielle ‚àÇf/‚àÇx = {df_dx}")
print(f" D√©riv√©e partielle ‚àÇf/‚àÇy = {df_dy}")

# Gradient (vecteur des d√©riv√©es partielles)
gradient = [df_dx, df_dy]
print(f" Gradient ‚àáf = {gradient}")

# √âvaluation en un point (1, 2)
point = {x: 1, y: 2}
gradient_point = [df.subs(point).evalf() for df in gradient]
print(f" Gradient en (1,2): {gradient_point}")

# D√©riv√©es partielles secondes
d2f_dx2 = diff(f_multi, x, 2)
d2f_dy2 = diff(f_multi, y, 2)
d2f_dxdy = diff(f_multi, x, y)
print(f"\n D√©riv√©e seconde ‚àÇ¬≤f/‚àÇx¬≤ = {d2f_dx2}")
print(f" D√©riv√©e seconde ‚àÇ¬≤f/‚àÇy¬≤ = {d2f_dy2}")
print(f" D√©riv√©e crois√©e ‚àÇ¬≤f/‚àÇx‚àÇy = {d2f_dxdy}")

# Matrice Hessienne
hessienne = sp.Matrix([
    [d2f_dx2, d2f_dxdy],
    [d2f_dxdy, d2f_dy2]
])
print(f"\n Matrice Hessienne:")
sp.pprint(hessienne)

# Test avec une autre fonction
print(f"\n TEST AVEC FONCTION QUADRATIQUE:")
f_quad = a*x**2 + b*x*y + c*y**2
print(f"f(x,y) = {f_quad}")
grad_quad = [diff(f_quad, x), diff(f_quad, y)]
print(f"Gradient: {grad_quad}")
```

## BLOC 7: Comparaison: D√©riv√©e Symbolique vs Num√©rique

```python
# BLOC 7: Comparaison symbolique vs num√©rique
print(" COMPARAISON: D√âRIV√âE SYMBOLIQUE VS NUM√âRIQUE")

# M√©thode symbolique (exacte)
f_sym = sin(x)*x
f_prime_sym = diff(f_sym, x)
print(f" D√©riv√©e symbolique exacte: {f_prime_sym}")

# √âvaluation exacte
x_eval = 1.0
exact_deriv = f_prime_sym.subs(x, x_eval).evalf()
print(f" f'({x_eval}) exact = {exact_deriv}")

# M√©thode num√©rique (approximation par diff√©rences finies)
def f_num(t):
    return t * np.sin(t)

h = 1e-6  # Pas tr√®s petit
approx_deriv = (f_num(x_eval + h) - f_num(x_eval)) / h
print(f" f'({x_eval}) approxim√© = {approx_deriv}")

# Calcul de l'erreur
erreur = abs(exact_deriv - approx_deriv)
print(f" Erreur d'approximation: {erreur}")
print(f" Pourcentage d'erreur: {erreur/abs(exact_deriv)*100:.6f}%")

# Test avec diff√©rentes valeurs de h
print("\n Test avec diff√©rents pas h:")
for h_test in [1e-4, 1e-6, 1e-8]:
    approx = (f_num(x_eval + h_test) - f_num(x_eval)) / h_test
    erreur_test = abs(exact_deriv - approx)
    print(f"h = {h_test:.0e} ‚Üí erreur = {erreur_test:.2e}")

# Test avec une fonction plus complexe
print(f"\n TEST AVEC FONCTION COMPLEXE:")
f_complex_sym = exp(-x/2) * sin(2*x)
f_complex_prime_sym = diff(f_complex_sym, x)
print(f"Fonction: {f_complex_sym}")
print(f"D√©riv√©e exacte: {f_complex_prime_sym}")

# Conversion pour calcul num√©rique
f_complex_num = sp.lambdify(x, f_complex_sym, 'numpy')
x_test_complex = 2.0
exact_complex = f_complex_prime_sym.subs(x, x_test_complex).evalf()
approx_complex = (f_complex_num(x_test_complex + h) - f_complex_num(x_test_complex)) / h
print(f"f'({x_test_complex}) exact = {exact_complex}")
print(f"f'({x_test_complex}) approx = {approx_complex}")
print(f"Erreur: {abs(exact_complex - approx_complex):.2e}")
```

## Conversion Symbolique ‚Üí Num√©rique avec Lambdify

```python
# BLOC 8 CORRIG√â: Conversion avec lambdify
print(" CONVERSION SYMBOLIQUE ‚Üí NUM√âRIQUE")

# Expression symbolique
f_sym = x**2 + sin(x)/x
print(f" Expression symbolique: {f_sym}")

# Conversion en fonction num√©rique
f_num = sp.lambdify(x, f_sym, 'numpy')
print(" Conversion r√©ussie avec lambdify")

# Test sur un point
test_point = 2.0
resultat_point = f_num(test_point)
print(f" Test en x={test_point}: {resultat_point}")

# Test sur un tableau (calcul vectoris√©)
x_array = np.linspace(0.1, 10, 5)  # √âviter x=0
y_array = f_num(x_array)
print(f" √âvaluation vectoris√©e:")
print(f"x = {x_array}")
print(f"f(x) = {y_array}")

# V√©rification avec calcul symbolique - CORRECTION ICI
print("\n V√©rification avec calcul symbolique:")
for x_val in x_array:
    sym_val = float(f_sym.subs(x, x_val).evalf())  # Conversion en float
    num_val = f_num(x_val)
    erreur = abs(sym_val - num_val)
    print(f"x={x_val:.1f}: symbolique={sym_val:.6f}, num√©rique={num_val:.6f}, erreur={erreur:.2e}")

# Test avec fonction multivari√©e
print(f"\n TEST FONCTION MULTIVARI√âE:")
f_multi_sym = x**2 + x*y + y**2
print(f"f(x,y) = {f_multi_sym}")

# Conversion pour deux variables
f_multi_num = sp.lambdify((x, y), f_multi_sym, 'numpy')
test_x, test_y = 2.0, 3.0
result_multi = f_multi_num(test_x, test_y)
print(f"f({test_x}, {test_y}) = {result_multi}")

# V√©rification symbolique - CORRECTION ICI
verif_multi = float(f_multi_sym.subs({x: test_x, y: test_y}).evalf())  # Conversion en float
print(f"V√©rification symbolique: {verif_multi}")
print(f"Erreur: {abs(verif_multi - result_multi):.2e}")

# Performance test
print(f"\n TEST DE PERFORMANCE:")
import time

x_large = np.linspace(1, 100, 10000)

# Temps calcul symbolique (point par point) - CORRECTION ICI
start_time = time.time()
results_sym = [float(f_sym.subs(x, val).evalf()) for val in x_large[:10]]  # Juste 10 points + conversion
time_sym = time.time() - start_time

# Temps calcul num√©rique (vectoris√©)
start_time = time.time()
results_num = f_num(x_large)
time_num = time.time() - start_time

print(f"Temps calcul symbolique (10 points): {time_sym:.4f}s")
print(f"Temps calcul num√©rique (10000 points): {time_num:.4f}s")
print(f"Acc√©l√©ration: {time_sym/(time_num/1000):.0f}x plus rapide")
```

## BLOC 10: Applications en Optimisation

```python
# BLOC 10 CORRIG√â 2: Applications en optimisation
print(" APPLICATIONS EN OPTIMISATION")

# Exemple 1: Optimisation d'une fonction simple
print("1. OPTIMISATION D'UNE FONCTION SIMPLE")
f_opt = x**3 - 6*x**2 + 9*x + 1
print(f" Fonction √† optimiser: f(x) = {f_opt}")

# D√©riv√©e premi√®re (condition n√©cessaire)
f_prime = diff(f_opt, x)
print(f" D√©riv√©e: f'(x) = {f_prime}")

# Points critiques
critical_points = solve(f_prime, x)
print(f" Points critiques: {critical_points}")

# D√©riv√©e seconde (test)
f_double_prime = diff(f_opt, x, 2)
print(f" D√©riv√©e seconde: f''(x) = {f_double_prime}")

# Classification des points critiques
print("\n CLASSIFICATION DES POINTS CRITIQUES:")
for point in critical_points:
    second_deriv = f_double_prime.subs(x, point)
    f_value = f_opt.subs(x, point)
    
    if second_deriv > 0:
        point_type = "minimum local"
    elif second_deriv < 0:
        point_type = "maximum local"
    else:
        point_type = "point d'inflexion"
    
    print(f"x = {point}: f(x) = {float(f_value):.2f}, {point_type}")

# Exemple 2: Optimisation multivari√©e
print(f"\n2. OPTIMISATION MULTIVARI√âE")
f_multi_opt = x**2 + x*y + y**2 - 4*x - 5*y
print(f" Fonction multivari√©e: f(x,y) = {f_multi_opt}")

# Gradient
gradient = [diff(f_multi_opt, var) for var in (x, y)]
print(f" Gradient:")
print(f"‚àÇf/‚àÇx = {gradient[0]}")
print(f"‚àÇf/‚àÇy = {gradient[1]}")

# Points critiques - CORRECTION ICI
critical_points_multi = solve(gradient, (x, y))
print(f" Points critiques: {critical_points_multi}")

# Matrice Hessienne
hessian = sp.Matrix([
    [diff(f_multi_opt, x, x), diff(f_multi_opt, x, y)],
    [diff(f_multi_opt, y, x), diff(f_multi_opt, y, y)]
])
print(f" Matrice Hessienne:")
sp.pprint(hessian)

# Classification des points critiques multivari√©s - CORRECTION COMPL√àTE
print(f"\n CLASSIFICATION MULTIVARI√âE:")
if critical_points_multi:
    # Si on a un seul point critique
    if isinstance(critical_points_multi, dict):
        point_dict = critical_points_multi
        x_val = point_dict[x]
        y_val = point_dict[y]
        
        # √âvaluation de la Hessienne au point critique
        hessian_point = hessian.subs(point_dict)
        det_hessian = hessian_point.det()
        f_value = f_multi_opt.subs(point_dict)
        
        if det_hessian > 0:
            if hessian_point[0, 0] > 0:
                point_type = "minimum local"
            else:
                point_type = "maximum local"
        elif det_hessian < 0:
            point_type = "point selle"
        else:
            point_type = "ind√©termin√©"
        
        print(f"Point ({x_val}, {y_val}): f = {float(f_value):.2f}, {point_type}")
    
    # Si on a plusieurs points critiques (liste de dict)
    elif isinstance(critical_points_multi, list):
        for point_dict in critical_points_multi:
            x_val = point_dict[x]
            y_val = point_dict[y]
            
            # √âvaluation de la Hessienne au point critique
            hessian_point = hessian.subs(point_dict)
            det_hessian = hessian_point.det()
            f_value = f_multi_opt.subs(point_dict)
            
            if det_hessian > 0:
                if hessian_point[0, 0] > 0:
                    point_type = "minimum local"
                else:
                    point_type = "maximum local"
            elif det_hessian < 0:
                point_type = "point selle"
            else:
                point_type = "ind√©termin√©"
            
            print(f"Point ({x_val}, {y_val}): f = {float(f_value):.2f}, {point_type}")
else:
    print("Aucun point critique trouv√©")

# Exemple 3: Optimisation avec contrainte (Lagrangien)
print(f"\n3. OPTIMISATION SOUS CONTRAINTE")
# Maximiser f(x,y) = x*y sous contrainte x + y = 10
f_obj = x * y
contrainte = x + y - 10

# Lagrangien
lambda_sym = symbols('lambda')
lagrangien = f_obj - lambda_sym * contrainte
print(f" Fonction objectif: {f_obj}")
print(f" Contrainte: {contrainte} = 0")
print(f" Lagrangien: L = {lagrangien}")

# Conditions KKT
grad_L = [diff(lagrangien, var) for var in (x, y, lambda_sym)]
print(f" Gradient du Lagrangien:")
for i, eq in enumerate(grad_L):
    print(f"‚àÇL/‚àÇ{['x', 'y', 'Œª'][i]} = {eq}")

# R√©solution
solution_kkt = solve(grad_L, (x, y, lambda_sym))
print(f" Solution KKT: {solution_kkt}")

# V√©rification
if solution_kkt:
    # G√©rer diff√©rents formats de solution
    if isinstance(solution_kkt, dict):
        x_opt = solution_kkt[x]
        y_opt = solution_kkt[y]
    else:
        x_opt, y_opt, lambda_opt = list(solution_kkt[0])[:2]
    
    f_opt_value = f_obj.subs({x: x_opt, y: y_opt})
    print(f" Valeur optimale: f({x_opt}, {y_opt}) = {f_opt_value}")

# Exemple 4: Application pratique - optimisation de volume
print(f"\n4. APPLICATION PRATIQUE: OPTIMISATION DE VOLUME")
print("Probl√®me: Trouver les dimensions d'une bo√Æte sans couvercle de volume maximal avec surface fixe")

# Surface fixe = 12 m¬≤
S = 12
# Dimensions: longueur = l, largeur = l, hauteur = h
l, h = symbols('l h', positive=True)

# Surface: fond (l¬≤) + 4 c√¥t√©s (l*h)
surface_eq = l**2 + 4*l*h - S
# Volume: V = l¬≤ * h
volume = l**2 * h

# Exprimer h en fonction de l depuis la contrainte de surface
h_expr = solve(surface_eq, h)[0]
print(f" Hauteur en fonction de la longueur: h = {h_expr}")

# Volume en fonction de l seulement
volume_l = volume.subs(h, h_expr)
print(f" Volume: V(l) = {volume_l}")

# Optimisation
dV_dl = diff(volume_l, l)
points_critiques = solve(dV_dl, l)
print(f" Points critiques: {points_critiques}")

# S√©lectionner la solution positive
l_opt = [sol for sol in points_critiques if sol > 0][0]
h_opt = h_expr.subs(l, l_opt)
volume_opt = volume_l.subs(l, l_opt)

print(f" Solution optimale:")
print(f"  Longueur = {float(l_opt):.2f} m")
print(f"  Hauteur = {float(h_opt):.2f} m") 
print(f"  Volume maximal = {float(volume_opt):.2f} m¬≥")

print(f"\n R√âSUM√â DES AVANTAGES SYMPY POUR L'OPTIMISATION:")
print("‚úì Calcul exact des gradients et hessiennes")
print("‚úì R√©solution analytique des conditions d'optimalit√©")
print("‚úì Classification automatique des points critiques")
print("‚úì Gestion des contraintes via Lagrangien")
print("‚úì Applications pratiques directes pour INF4127")
```
